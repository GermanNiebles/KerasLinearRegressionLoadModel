```
### Importar tensorflow
import tensorflow as tf
```

```
### Importar pandas
import numpy as np
```

```
### Importar keras de tensorflow
from tensorflow import keras
```

```
 ### Definir numero de neuronas
 numNeuronas = 1
```

```
#### Instanciacion del modelo sequential
### el shape de entrada es de uno
model = keras.Sequential([keras.layers.Dense(units=numNeuronas, input_shape=[1])])
```

```
### Path de la ubicacion del modelo
model_save_path = "./LRDeepLearning"
```

```
### Cargar el modelo
load_model = tf.saved_model.load(model_save_path)
```

```
### Predecir
print(model.predict([10.0]))
```

```
### Pendiente de conocer
print(list(load_model.signatures.keys()))
```

```
### Pendiente de conocer
infer = load_model.signatures['serving_default']
```

```
### Cargar resultado de prediccion
predictions = model.predict([10])
```

```
#### Aplicar softmax para ver el score
score = tf.nn.softmax(predictions[0])
```

```
### Pendiente de conocer
np.argmax(score)
```
